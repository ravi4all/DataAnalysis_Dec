```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('data/ipl/matches.csv')
```

```python
dataset.head()
```

# Matches Per Season

```python
# plt.hist(dataset['season'])
# plt.show()

sns.countplot(x = 'season', data = dataset)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x = 'winner', data = dataset)
plt.xticks(rotation = 'vertical')
plt.show()
```

# Number of Matches Played By Each Team

```python
temp_df.head()
```

```python
temp_df.tail()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x = 'value', data = temp_df)
plt.xticks(rotation = 'vertical')
plt.show()
```

# Toss Decision

```python
sns.countplot(x = 'toss_decision', data = dataset)
plt.show()
```

```python
temp_df = dataset.toss_decision.value_counts()
```

```python
temp_df
```

```python
import numpy as np
```

```python
labels = np.array(temp_df.index)
```

```python
labels
```

```python
size =(np.array(temp_df / temp_df.sum() * 100))
```

```python
size
```

```python
color = ['gold', 'green']
plt.pie(size, labels=labels, colors=color, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Toss Decision')
plt.show()
```

# How Toss Decision Varied Over Time

```python
plt.figure(figsize=(12,6))
sns.countplot(x = 'season', hue = 'toss_decision', data=dataset)
plt.show()
```
